## Broadcasting
### Question 10: Company X wants to calculate the profit margin percentage for each month using the given revenue data array and a constant cost value of 4000 Rs. Apply broadcasting to perform this calculation.

<details>
    <summary> Hint </summary>
    
```markdown
Broadcasting allows us to perform element-wise operations between arrays of different shapes. Here, the revenue data array is subtracted by the cost value, and then divided by the revenue data array. The result is multiplied by 100 to obtain the profit margin percentage for each month.

```
</details>

<details>
    <summary> Solution </summary>

```python

cost = 4000
profit_margin_percentage = ((revenue_data - cost) / revenue_data) * 100
print(profit_margin_percentage)

```
</details>
